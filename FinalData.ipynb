```python
match_df['Team1'] = match_df['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['Team2'] = match_df['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['Team1'] = match_df['Team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['Team2'] = match_df['Team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
delivery_df = match_df.merge(delivery,on='ID')
```

```python
delivery_df['WinningTeam'] = delivery_df['WinningTeam'].str.replace('Delhi Daredevils','Delhi Capitals')
delivery_df['BattingTeam'] = delivery_df['BattingTeam'].str.replace('Delhi Daredevils','Delhi Capitals')
delivery_df['WinningTeam'] = delivery_df['WinningTeam'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
delivery_df['BattingTeam'] = delivery_df['BattingTeam'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
delivery_df = delivery_df[delivery_df['innings'] == 2]
```

```python
delivery_df.head()
```

```python
def get_bowling_team(row):
    if row['WinningTeam'] == row['Team1'] and row['WinningTeam'] == row['BattingTeam']:
        return row['Team2']
    elif row['WinningTeam'] == row['Team2'] and row['WinningTeam'] == row['BattingTeam']:
        return row['Team1']
    elif row['WinningTeam'] == row['Team1'] and row['Team2'] == row['BattingTeam']:
        return row['WinningTeam']
    elif row['WinningTeam'] == row['Team2'] and row['Team1'] == row['BattingTeam']:
        return row['WinningTeam']
```

```python
delivery_df['BowlingTeam'] = delivery_df.apply(get_bowling_team, axis=1)
```

```python
delivery_df.tail()
```

```python
delivery_df['current_score'] = delivery_df.groupby('ID').cumsum()['total_run_y']
```

```python
delivery_df
```

```python
delivery_df['runs_left'] = delivery_df['total_run_x'] - delivery_df['current_score'] + 1
```

```python
delivery_df
```

```python
delivery_df['balls_left'] = 120 - (delivery_df['overs']*6 + delivery_df['ballnumber'])
```

```python
delivery_df
```

```python
delivery_df['player_out'] = delivery_df['player_out'].fillna("0")
delivery_df['player_out'] = delivery_df['player_out'].apply(lambda x:x if x == "0" else "1")
delivery_df['player_out'] = delivery_df['player_out'].astype('int')
wickets = delivery_df.groupby('ID').cumsum()['player_out'].values
delivery_df['wickets'] = 10 - wickets
delivery_df.head()
```

```python
delivery_df['crr'] = (delivery_df['current_score'] *6) / (120 - delivery_df['balls_left'])
```

```python
delivery_df['rrr'] = (delivery_df['runs_left'] *6)/ delivery_df['balls_left']
```

```python
def result(row):
    return 1 if row['BattingTeam'] == row['WinningTeam'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis = 1)
```

```python
delivery_df
```

```python
final_df = delivery_df[['BattingTeam','BowlingTeam','City','runs_left','balls_left','wickets','total_run_x','crr','rrr','result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

# Model Building

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
x = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['BattingTeam','BowlingTeam','City'])
], remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
